```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('world_Data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data2 = data.melt(
    id_vars= ["Country Name", "Country Code", "Series Name", "Series Code"],
    var_name= "Year_label",
    value_name= "Value"
)
```

```python
data2.info()
```

```python
data2.head()
```

```python
data2['Year'] = data2['Year_label'].str[:4]
```

```python
data2.head(2)
```

```python
data2['Year'] = data2['Year'].astype(int)
```

```python
data2['Value'] = data2['Value'].replace("..", np.nan)
```

```python
data2['Value'] = data2['Value'].astype(float)
```

```python
data2.head(2)
```

```python
data2.to_csv('world_Data_unpivoted.csv')
```
